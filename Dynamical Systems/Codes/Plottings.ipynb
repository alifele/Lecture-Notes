```python
xList = np.linspace(-1*np.pi,1*np.pi,50)
yList = np.linspace(-np.pi,np.pi,50)
X,Y = np.meshgrid(xList, yList)

def L(X,Y):
    return 1/2*Y**2 - np.cos(X)

def F(X,Y,d):
    return [Y,-d*Y-np.sin(X)]
```

```python
U,V = F(X,Y,1)
plt.streamplot(X,Y,U,V)
plt.contour(X,Y,L(X,Y), levels=[0],linewidths=3)
#lt.savefig("../Images/LyapunovExample.pdf", format="pdf", bbox_inches="tight")
```
